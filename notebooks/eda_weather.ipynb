```python
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt
import os

DATA_PATH = Path(r"C:\Users\safar\OneDrive\Рабочий стол\elbrus\my_project_2_1\data\weather")


# Посмотрим, какие есть классы
classes = [d.name for d in DATA_PATH.iterdir() if d.is_dir()]
print("Классы:", classes)

# Покажем по 1 изображению из каждого класса
fig, axs = plt.subplots(1, len(classes), figsize=(15, 5))
for i, cls in enumerate(classes):
    img_path = next((DATA_PATH / cls).glob("*.jpg"))
    img = Image.open(img_path)
    axs[i].imshow(img)
    axs[i].set_title(cls)
    axs[i].axis('off')

plt.show()
```
